```python
import json, uuid
from jinja2 import Template
from IPython.display import display, HTML

# Define the advanced Jinja template.
template_str = """
<!DOCTYPE html>
<html>
<head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width,initial-scale=1.0,maximum-scale=1.0,minimum-scale=1.0" />
    <title>{{ page_title }}</title>
    <!-- Adding Google Font -->
    <link rel="preconnect" href="https://fonts.googleapis.com" />
    <link rel="preconnect" href="https://fonts.gstatic.com" crossorigin />
    <link href="https://fonts.googleapis.com/css2?family=Roboto&display=swap" rel="stylesheet" />
    <!-- Adding the standalone version of Lightweight Charts and Axios -->
    <script type="text/javascript" src="{{ script_src }}"></script>
    <script type="text/javascript" src="{{ axios_src }}"></script>
    <style>
        body {
            padding: 0;
            margin: 0;
            background-color: #222;
            font-family: 'Roboto', sans-serif;
        }
        /* Ensure container occupies full viewport */
        #{{ output_div }} {
            position: absolute;
            width: 100%;
            height: 100%;
        }
    </style>
</head>
<body>
   <div id="{{ output_div }}"></div>
   <script type="text/javascript">
     (() => {
         // Get the container by its unique id.
         const container = document.getElementById("{{ output_div }}");
         // Create the chart with provided options.
         const chart = LightweightCharts.createChart(container, {{ chart.options }});
         this.chart = chart;
         
         // Create a legend element to display dynamic info.
         const legend = document.createElement('div');
         legend.style = "position: absolute; left: 10px; top: 10px; z-index: 1; font-size: 10px; font-family: sans-serif; line-height: 14px; font-weight: 200;";
         container.appendChild(legend);
         
         // Dynamically add each chart series.
         {% for series in chart.series %}
         (() => {
             // Create a row for this series in the legend.
             const row_{{ series.series_name }} = document.createElement('div');
             row_{{ series.series_name }}.innerHTML = '<a href="https://tradingview.github.io/lightweight-charts/">Made By DrJuneMoone</a>';
             row_{{ series.series_name }}.style.color = 'orange';
             legend.appendChild(row_{{ series.series_name }});
             
             // Add the series to the chart.
             const chart_series_{{ series.series_name }} = chart.addSeries(LightweightCharts.{{ series.series_type }}Series, {{ series.options }});
             chart_series_{{ series.series_name }}.setData({{ series.data }});
             chart_series_{{ series.series_name }}.setMarkers({{ series.markers }});
             
             // Add price lines if any.
             {% for price_line in series.price_lines %}
             chart_series_{{ series.series_name }}.createPriceLine({{ price_line }});
             {% endfor %}
             
             // Expose the series instance for future updates.
             this.chart_series_{{ series.series_name }} = chart_series_{{ series.series_name }};
             
             // Subscribe to click events.
             chart.subscribeClick(function (param) {
                 console.log(`User clicked at (${param.point.x}, ${param.point.y}) at time ${param.time}`);
             });
             
             // Subscribe to crosshair movement to update the legend.
             chart.subscribeCrosshairMove(param => {
                 if (!param.point) {
                     return;
                 }
                 if (param.time) {
                     const data = param.seriesData.get(chart_series_{{ series.series_name }});
                     {% if series.series_name == 'ohlc' %}
                     const open = data.open !== undefined ? data.open : data.value;
                     const high = data.high !== undefined ? data.high : data.value;
                     const low = data.low !== undefined ? data.low : data.value;
                     const close = data.close !== undefined ? data.close : data.value;
                     row_{{ series.series_name }}.innerHTML = `<strong>open: ${open.toFixed(2)} | high: ${high.toFixed(2)} | low: ${low.toFixed(2)} | close: ${close.toFixed(2)}</strong>`;
                     {% else %}
                     const price = data.value !== undefined ? data.value : data.close;
                     row_{{ series.series_name }}.innerHTML = `<strong>{{ series.series_name }}: ${price.toFixed(2)}</strong>`;
                     {% endif %}
                 } else {
                     console.log(`User moved crosshair to (${param.point.x}, ${param.point.y}) at time ${param.time}`);
                 }
             });
             chart.timeScale().fitContent();
         })();
         {% endfor %}
         
         // Resize the chart when the window size changes.
         window.addEventListener("resize", () => {
             this.chart.resize(window.innerWidth, window.innerHeight);
         });
     })();
     
     // Data update logic using Axios.
     const data_url = "{{ data_url }}";
     this.data_url = data_url;
     function updateData() {
         try {
             return axios({
                 method: 'GET',
                 url: data_url,
                 crossOrigin: '*',
             })
             .then(response => {
                 {% for series in chart.series %}
                 // Assume response.data contains an array for each series.
                 this.chart_series_{{ series.series_name }}.update(response.data.{{ series.series_name }}[0]);
                 {% endfor %}
             });
         } catch (error) {
             console.error("Data update error:", error);
         }
     }
     
     function setDataUpdateInterval() {
         var currentDateSeconds = new Date().getSeconds();
         if (currentDateSeconds === 0) {
             setInterval(updateData, 60000);
         } else {
             setTimeout(() => {
                 setDataUpdateInterval();
             }, (60 - currentDateSeconds) * 1000);
         }
         updateData();
     }
     setDataUpdateInterval();
   </script>
</body>
</html>
"""

# Create a Template object.
template = Template(template_str)

# Define chart options.
chart_options = {
    "layout": {"backgroundColor": "#222", "textColor": "#d1d4dc"},
    "timeScale": {"timeVisible": True, "secondsVisible": False},
}
# For demonstration, define one series. You can add more to the list.
chart_series = [
    {
        "series_name": "ohlc",                  # Unique name for the series.
        "series_type": "Candlestick",           # Series type (e.g., Candlestick, Line, Area).
        "options": json.dumps({"upColor": "#26a69a", "downColor": "#ef5350", "borderVisible": False, "wickVisible": True}),
        "data": json.dumps([
            {"time": "2019-05-23", "open": 188.45, "high": 192.54, "low": 186.27, "close": 192.0},
            {"time": "2019-05-24", "open": 192.54, "high": 193.86, "low": 190.41, "close": 193.59}
        ]),
        "markers": json.dumps([]),
        "price_lines": []                       # Add any price line options if needed.
    }
]

chart_config = {
    "options": json.dumps(chart_options),
    "series": chart_series,
}

# Other configurations.
page_title = "Lightweight Charts Customization Tutorial"
script_src = "https://unpkg.com/lightweight-charts/dist/lightweight-charts.standalone.production.js"
axios_src = "https://unpkg.com/axios/dist/axios.min.js"
data_url = "https://example.com/data"  # Replace with your actual data endpoint.
output_div = "chart-" + uuid.uuid4().hex

# Render the template with dynamic variables.
rendered_html = template.render(
    page_title=page_title,
    script_src=script_src,
    axios_src=axios_src,
    data_url=data_url,
    output_div=output_div,
    chart=chart_config
)

display(HTML(rendered_html))
```
